```python
from bs4 import BeautifulSoup
from urllib import request
import requests
import pandas as pd
import numpy as np
```

```python
url = "https://www.boomplay.com/playlists/880218?from=home"
r = requests.get(url)
```

```python
r.text[0:500]
```

```python
soup = BeautifulSoup(r.text, "html.parser")
```

```python
# top 100 tracks in africa
songs= soup.find_all("li", attrs= {"class": "clearfix play_one"})
len(songs)
```

```python
# finding the song and artist
song = songs[0]
song1 = song.find_all("a")
song1
```

```python
# looping through all songs
records = []
for i in songs:
    song_info = i.find_all("a")
    song = song_info[0].text
    artist = song_info[1].text
    print(artist)
    print(song)
    print()
    records.append((song, artist))
```

```python
records[0:3]
```

```python
# putting the info in a dataframe
df = pd.DataFrame(records, columns = ["song name", "artist"])
df
```

Top 5 artists with most songs on top 100 in Africa

```python
rank = df.groupby("artist")["song name"].agg(["count"]).sort_values("count", ascending = False)
rank.head(5)
```
